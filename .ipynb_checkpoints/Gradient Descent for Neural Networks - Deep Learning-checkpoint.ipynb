# Gradient Descent for Neural Networks
## Implementing gradient descent for neural networks (or logistic regression)


#### Predicting if a person would buy life insurance based on his age using logistic regression

Above is a binary logistic regression problem as there are only two possible outcomes (i.e. if person buys insurance or not)

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Implementing neural network using Tensorflow

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['age', 'affordibility']], df.bought_insurance, test_size=0.2, random_state=2)
```

```python
# Scaling is done to bring values between 0 and 1 for classification
x_train_scaled = x_train.copy()
x_train_scaled['age'] = x_train_scaled['age'] / 100

x_test_scaled = x_test.copy()
x_test_scaled['age'] = x_test_scaled['age'] / 100
```

```python
model = keras.Sequential([
        keras.layers.Dense(1,                            # Output neuron
                           input_shape=(2,),             # 2-input neuron
                           activation='sigmoid',         # Activation function
                           kernel_initializer='ones',    # Initial weights
                           bias_initializer='zeros')     # Initial biases
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train_scaled, y_train, epochs=5000)
```

```python
model.evaluate(x_test_scaled, y_test)
```

#### Instead of model.predict, we will write our own prediction function that uses w1,w2 and bias

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

sigmoid(10)
```

```python
def prediction_function(age, affordibility):
    # weighted_sum(y) = weight01 * input01 + weight02 * input02 + bias
    weighted_sum = coef[0] * age + coef[1] * affordibility + intercept
    
    # prediction
    return sigmoid(weighted_sum)

prediction_function(.25, 0)
```

#### Implementing neural network from scratch

**Now we start implementing gradient descent in plain python. Again the goal is to come up with same w1, w2 and bias that keras model calculated. We want to show how keras/tensorflow would have computed these values internally using gradient descent**

**First write couple of helper routines such as sigmoid and log_loss**

```python
# Making sigmoid func for numpy arrays
def sigmoid_numpy(X):
    return 1 / (1 + np.exp(-X))

sigmoid_numpy(np.array([12, 0, 1]))
```

```python
# It is advised to use log loss / binary cross-entropy with logistic regression
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i, epsilon) for i in y_predicted]
    y_predicted_new = [min(i, 1 - epsilon) for i in y_predicted_new]
    
    y_predicted_new = np.array(y_predicted_new)
    return -(np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * (np.log(1 - y_predicted_new))))
```

#### Formul to calculate derivatives


#### All right now comes the time to implement our final gradient descent function

```python
def gradient_descent(age, affordibility, y_true, epochs, threshold):
    # Initializing weights and bias
    w1 = w2 = 1
    b = 0
    
    # Initializing learning rate (0.01 is used as convention)
    learning_rate = 0.5
    
    # Initializing total values in data set
    n = len(age)
    
    for i in range(epochs):
        # Calculating weighted_sum(y)
        weighted_sum = w1 * age + w2 * affordibility + b
        
        # Prediciting value (using sigmoid)
        y_predicted = sigmoid_numpy(weighted_sum)
        
        # Calculating loss
        loss = log_loss(y_true, y_predicted)
        
        # Calculating derivatives
        w1d = (1 / n) * np.dot(np.transpose(age), (y_predicted - y_true))
        w2d = (1 / n) * np.dot(np.transpose(affordibility), (y_predicted - y_true))
        bias_d = np.mean(y_predicted - y_true)
        
        # Adjusting weights and bias
        w1 = w1 - learning_rate * w1d
        w1 = w2 - learning_rate * w2d
        b = b - learning_rate * bias_d
        
        print(f'Epoch: {i}, w1: {w1}, w2: {w2}, bias: {b}, loss: {loss}')
        
        if loss <= threshold:
            break
        
    return w1, w2, b
```

```python
def gradient_descent_new(age, affordability, y_true, epochs, loss_thresold):
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)
    for i in range(epochs):
        weighted_sum = w1 * age + w2 * affordability + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true, y_predicted)

        w1d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true)) 
        w2d = (1/n)*np.dot(np.transpose(affordability),(y_predicted-y_true)) 

        bias_d = np.mean(y_predicted-y_true)
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        bias = bias - rate * bias_d

        print (f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')

        if loss<=loss_thresold:
            break

    return w1, w2, bias
```

```python
gradient_descent_new(x_train_scaled['age'], x_train_scaled['affordibility'], y_train, 5000, 0.4781)
```
